```python
def read_by_cpg_list(df_loc,cpg_list):
    df=pd.read_csv(df_loc, index_col=0,low_memory=False)
    col_p=list(df.columns[1:])
    drop_p=[]
    for i in col_p:
        if i not in cpg_list:
            drop_p.append(i)

    df=df.drop(columns=drop_p)    
    return df
```
