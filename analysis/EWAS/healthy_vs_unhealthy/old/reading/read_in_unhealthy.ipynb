# Reading in brain data - just unhealthy

```python
import pandas as pd
import _pickle as cPickle
import numpy as np
```

### Healthy IDs

```python
with open(r"../healthy_ids_from_unhealthy", "rb") as input_file:
    healthy_ids = cPickle.load(input_file)
```

### Alzheimer's

```python
alzheimers_df=pd.read_csv('../brain/alzheimers/alz_brain_all.csv',index_col=0)  
```

```python
alzheimers_df_ids=list(alzheimers_df.index)
for i in alzheimers_df_ids:
    if i in healthy_ids:
        alzheimers_df=alzheimers_df.drop([i], axis=0)
alzheimers_df
```

```python
alzheimers_df
```

```python
unhealthy_meta=pd.read_csv('../unhealthy_meta.csv',index_col=0)  
```

```python
alzheimers_df_ids=list(alzheimers_df.index)
unhealthy_meta_ids=list(unhealthy_meta.index)
alzheimers_df['disease']=np.zeros(len(alzheimers_df_ids))
alzheimers_df['tissue']=np.zeros(len(alzheimers_df_ids))

for i in alzheimers_df_ids:
    alzheimers_df.loc[i,'disease']=unhealthy_meta.loc[i,'disease']
    alzheimers_df.loc[i,'tissue']=unhealthy_meta.loc[i,'tissue']
alzheimers_df
```

```python
cols = list(alzheimers_df.columns.values)
cols=cols[3:]
alzheimers_df = alzheimers_df[['AGE','disease','tissue']+cols]
alzheimers_df
```

```python
alzheimers_df.to_csv('../brain/alzheimers/alz_brain_unhealthy_all.csv', encoding='utf-8', index=True)
```

### Huntington's

```python
huntingtons_df=pd.read_csv('../brain/huntingtons/hunt_brain_all.csv',index_col=0)  
```

```python
huntingtons_df_ids=list(huntingtons_df.index)
for i in huntingtons_df_ids:
    if i in healthy_ids:
        huntingtons_df=huntingtons_df.drop([i], axis=0)
huntingtons_df
```

```python
huntingtons_df=huntingtons_df.drop(['sample_type'], axis=1)
huntingtons_df
```

```python
huntingtons_df.to_csv('../brain/huntingtons/hunt_brain_unhealthy_all.csv', encoding='utf-8', index=True)
```
