```python
import csv
import sqlite3
import os
import numpy as np
import pandas as pd

```

```python
#Opening patient statusinfor
with open('Patient_Status.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
```

```python
#No of HC patients
healthy_patient=[]
for i in range(len(data)):
    if data[i][1]=='HC':
        healthy_patient.append(int(data[i][0]))
        
len(healthy_patient)
```

```python
#Opening blood_chem_data
with open('Blood_Chemistry___Hematology.csv', newline='') as csvfile:
    blood_data = list(csv.reader(csvfile))
```

```python
#List of blood tests
test=[]
for i in range(1,34):
    test.append(blood_data[i][12])
test
```

```python
#Patient IDs for healthy and unhealty groups
healthy_blood_data=[]
unique_healthy_db=[]
unique_unhealthy_db=[]
for i in range(1,len(blood_data)):
    if int(blood_data[i][0]) in healthy_patient:
        healthy_blood_data.append(blood_data[i])
        if int(blood_data[i][0]) not in unique_healthy_db:
            unique_healthy_db.append(int(blood_data[i][0]))
    else:
        if int(blood_data[i][0]) not in unique_unhealthy_db:
            unique_unhealthy_db.append(int(blood_data[i][0]))
```

```python
healthy_blood_data
len(unique_healthy_db)
```

```python
unique_bd_patient_ids=unique_unhealthy_db+unique_healthy_db
```

```python
data = pd.read_csv ('Blood_Chemistry___Hematology.csv')   
blood_chem_df = pd.DataFrame(data)

def filter_rows_by_values(df, col, values):
    return df[df[col].isin(values) == True]

#blood_chem_df=filter_rows_by_values(blood_chem_df,"PATNO", healthy_patient)
blood_chem_df=blood_chem_df.set_index("PATNO")
blood_chem_df
```

```python
screen_dem_data = pd.read_csv ('Screening___Demographics.csv')   
screen_dem_df = pd.DataFrame(screen_dem_data)
    
screen_dem_df=screen_dem_df.set_index('PATNO')
```

## Extracting data for EDA

```python
clean_blood_df = pd.DataFrame({"PATNO":unique_bd_patient_ids})
clean_blood_df=clean_blood_df.set_index('PATNO')
total=len(unique_bd_patient_ids)
clean_blood_df['YOB']=np.zeros(total)
clean_blood_df['AGE']=np.zeros(total)
for i in unique_bd_patient_ids:
    try:
        clean_blood_df.loc[i]['YOB']=screen_dem_df.loc[i]['BIRTHDT']
        clean_blood_df.loc[i]['AGE']=2010-screen_dem_df.loc[i]['BIRTHDT']
    except KeyError:
        clean_blood_df.loc[i]['YOB']="NaN"
        clean_blood_df.loc[i]['AGE']="NaN"
    
```

```python
clean_blood_df['HEALTHY']=np.zeros(total)

for i in unique_bd_patient_ids:
    if i in unique_healthy_db:
        clean_blood_df.loc[i]['HEALTHY']=1
    else:
        clean_blood_df.loc[i]['HEALTHY']=0
```

```python
for j in test:
    clean_blood_df[j]=np.zeros(total)
```

```python
for i in unique_bd_patient_ids:

    tmp=pd.DataFrame(blood_chem_df.loc[i])
    tmp=tmp.set_index('LTSTNAME')
    

    for j in test:
        try:
            v=tmp.loc[j]['LSIRES']
            try:
                v.dtype
                clean_blood_df.loc[i][j]=v[-1]
            except AttributeError:
                clean_blood_df.loc[i][j]=v
        except KeyError:
            clean_blood_df.loc[i][j]="N/A"
        
```
